```python
# 1) Input: [“baseball”, “b,bas,ba,all,base,seb,ball”]
# Output: base,ball
# Split the first input based on the longest possibility match in second input

ip = ["baseball", "b,bas,ba,all,base,seb,ball"]

ip1 = ip[1].split(",")
#print(ip1)
l = len(ip[0])

sub = []
index = 0
for i in range(len(ip[0])):
    if ip[0][:i] in ip1:
        sub.append(ip[0][:i])
        index = i
       
s1 = ""
sub1 = []
for j in ip[0][index:]:
    s1 += j
    if s1 in ip1:
        sub1.append(s1)
        
res =(max(sub),max(sub1))   
print(res)

  
    


```

```python
# 2) Write two best scenarios where inheritance can be used. Note:  don’t simply say that  
#     we can call parent class methods and attributes.

'''
scenario1
---------
- In e-commernce application there are multiple products are their 
- If you take all the products have some attributes are commom like name,cost_price,discount,manufacture etc.,
- This one taken as Parent class as Product class.
- If you take electronic product items have the parent class attributes and also add some additional attributes like color,
  warrenty_details etc., this one taken as Child class as ElectronicProduct class

'''
class Product:
    def __init__(self,name,cost_price,discount,manufacture):
        self.name=name
        self.cost_price =cost_price
        self.discount = discount
        self.manufacture = manufacture
    
    
class ElectronicProduct(Product):
    def __init__(self,color,warrenty_details):
        self.color = color
        self.warrenty_details = warrenty_details
        super().__init__()
    
    def get_product(self):
        return self.__dict__()
    
# p = Product("Iphone",123422,5000,"China")
# ep = ElectronicProduct("Iphone",123422,5000,"China","red", "1year",)

'''
scenario2
---------
- In different project if you using the customer application to your present application
- You need to add all the fields to your existing application along with you need to add addditional field to your product
- In this scenario you take the inheritance concept in which take the parent class of already existing class to your present
  class 
'''

class Customer:
    def __init__(self,name,email,phone,address):
        self.name=name
        self.email = email
        self.phone = phone
        self.address = address
    
    
class Add(Customer):
    def __init__(self,pancard,passport):
        self.pancard = pancard
        self.passport = passport 
        super().__init__()
    
    def get_customer(self):
        return self.__dict__()

e = Add("john", "john@gmail.com", 8787905642, "india")
```

```python
# 6) Define class can be create only one object. If the user create second object then it should return the first object

class Singleton:
    instance = None
    def __new__(cls):
        if cls.instance is None:
            cls.instance = super().__new__(cls)
        return cls.instance
    
obj1 = Singleton()
obj2 = Singleton()
print(obj1 is obj2)
```
